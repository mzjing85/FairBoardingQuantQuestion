```python
import pandas as pd
import matplotlib.pyplot as plt

df_prob = pd.read_excel("passenger_probabilities.xlsx")

plt.figure(figsize=(10,6))
plt.scatter(df_prob['NumPassengers'], df_prob['Probability'], s=10, alpha=0.7)
plt.title('Passenger Probability by Number of Passengers')
plt.xlabel('Number of Passengers')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
# Load your data
df_time = pd.read_excel("generatedTestData/elapsed_times.xlsx")

# Plot
plt.figure(figsize=(10,6))
plt.plot(df_time['NumPassengers'], df_time['AvgElapsedTime_ms'], marker='o')
plt.title('Average Elapsed Time per Passenger')
plt.xlabel('Number of Passengers')
plt.ylabel('Average Elapsed Time (seconds)')
plt.grid(True)
plt.show()
```
